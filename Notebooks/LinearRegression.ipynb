```python
import pandas as pd
```

```python
df=pd.read_csv('vehicles.csv')
```

```python
print(df.head())
```

```python
df=df.drop(['id','url','region_url','image_url','county','VIN','description','size'],axis=1)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    print(f"{col}: ")
    print(df[col].unique())
```

```python
mod_cols={
    'manufacturer', 'model', 'condition', 'cylinders',
    'fuel', 'title_status', 'transmission',
    'drive', 'type', 'paint_color'
}
for col in mod_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
med_cols={
    'year', 'odometer', 'lat', 'long'
}
for col in med_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
df=df.dropna(subset=['posting_date'])
df['posting_date']=pd.to_datetime(df['posting_date'], utc=True)
df['post_year']=df['posting_date'].dt.year
df['post_month']=df['posting_date'].dt.month
df.drop(columns=['posting_date'], inplace=True)
```

```python
df.isnull().sum()
```

```python
cols=['region','manufacturer','model','condition',
        'cylinders','fuel','title_status',
        'transmission','drive','type',
        'paint_color','state'
     ]
for col in cols:
    print(col, df[col].nunique())
```

```python
df.drop(columns=['model'], inplace=True)
```

```python
df=pd.get_dummies(
    df,
    columns=[
        'region','manufacturer','condition',
        'cylinders','fuel','title_status',
        'transmission','drive','type',
        'paint_color','state'
    ],
    drop_first=True
)
```

```python
df.isnull().sum()
```
